```python
import requests
import pandas as pd
url = 'https://github.com/vimmiseth/PySparkPracs/blob/main/Assessment.xlsx'

response = requests.get(url)

df2 = pd.ExcelFile('https://github.com/vimmiseth/PySparkPracs/blob/main/Assessment.xlsx?raw=true')
print(df2.sheet_names)
```

```python
xls = 'https://github.com/vimmiseth/PySparkPracs/blob/main/Assessment.xlsx?raw=true'
dft = pd. read_excel(xls, sheet_name = 'teacher')
dfs = pd. read_excel(xls, sheet_name ='Student')
dft.head()
```

```python
dfs.head()
```

```python
import pyspark
from pyspark.sql import SparkSession
#Create PySpark SparkSession
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("SparkByExamples.com") \
    .getOrCreate()
#Create PySpark DataFrame from Pandas

sdf_st=spark.createDataFrame(dfs) 
sdf_tr=spark.createDataFrame(dft) 
sdf_st.printSchema()
sdf_st.show()
sdf_tr.show()
```
